Implement and Visualize: 6 Kernel Functions

I(x) = 1, -1 <= x <= 1 or  = 0, rỗng

**6 Kernel**

1.   Triangular Kernel
2.   Rectangular (Uniform) Kernel
3.   Parabolic (Epanechnikov) Kernel
4.   Gaussian Kernel
5.   Silverman Kernel
6.   Biweight Kernel



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Danh sách tên Kernel
kernel_names = ["Triangular", "Rectangular", "Parabolic", "Gaussian", "Silverman", "Biweight"]

# Định nghĩa các kernel functions
def triangular_kernel(u):
    return (1 - np.abs(u)) * (np.abs(u) <= 1)

def rectangular_kernel(u):
    return 0.5 * (np.abs(u) <= 1)

def parabolic_kernel(u):
    return (3/4) * (1 - u**2) * (np.abs(u) <= 1)

def gaussian_kernel(u):
    return (1/np.sqrt(2*np.pi)) * np.exp(-0.5 * u**2)

def silverman_kernel(u):
    return 0.5 * np.exp(-np.abs(u) / np.sqrt(2)) * np.sin(np.abs(u) / np.sqrt(2) + np.pi / 4)

def biweight_kernel(u):
    return (15/16) * (1 - u**2)**2 * (np.abs(u) <= 1)
```

```python
# Tạo dữ liệu cho trục x
u = np.linspace(-3, 3, 1000000)
```

```python
# Ánh xạ tên kernel với hàm tương ứng
kernel_functions = {
    "Triangular": triangular_kernel,
    "Rectangular": rectangular_kernel,
    "Parabolic": parabolic_kernel,
    "Gaussian": gaussian_kernel,
    "Silverman": silverman_kernel,
    "Biweight": biweight_kernel
}
```

```python
# Vẽ đồ thị
plt.figure(figsize=(10, 6))
colors = ["orange", "blue", "green", "brown", "pink", "purple"]

for name, color in zip(kernel_names, colors):
    plt.plot(u, kernel_functions[name](u), label=name, color=color)

# Thiết lập nhãn và tiêu đề
plt.xlabel("u")
plt.ylabel("Kernel value")
plt.title("Kernel Functions")
plt.legend()
plt.grid()

# Hiển thị đồ thị
plt.show()
```
